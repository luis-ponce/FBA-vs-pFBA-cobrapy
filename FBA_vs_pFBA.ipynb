```python
biomassr_FBA = []
lacdflux_FBA = []
maxtca_FBA = []
mintca_FBA = []
co2flux_FBA = []

for i in range(-200, 200, 2):
    model = cobra.io.load_json_model("iYS854_reduced.json")
    set_boundaries(model)
    model.reactions.EX_lac__L_e.bounds = (i, i+ 1 )
#     solution = cobra.flux_analysis.pfba(model)
    solution = model.optimize()
    biomassr_FBA.append(solution.fluxes.BIOMASS_iYS_wild_type)
    lacdflux_FBA.append(solution.fluxes.EX_lac__L_e)
    co2flux_FBA.append(solution.fluxes.EX_co2_e)
    maxtca_FBA.append(max(np.abs(solution.fluxes.PDH) \
        , np.abs(solution.fluxes.OSUCCL) , np.abs(solution.fluxes.AKGDH) , np.abs(solution.fluxes.SUCOAS) \
        , np.abs(solution.fluxes.SUCD2) , np.abs(solution.fluxes.MDH3_1) , np.abs(solution.fluxes.FUM) \
        , np.abs(solution.fluxes.CS)))
    mintca_FBA.append(min(np.abs(solution.fluxes.PDH) \
        , np.abs(solution.fluxes.OSUCCL) , np.abs(solution.fluxes.AKGDH) , np.abs(solution.fluxes.SUCOAS) \
        , np.abs(solution.fluxes.SUCD2) , np.abs(solution.fluxes.MDH3_1) , np.abs(solution.fluxes.FUM) \
        , np.abs(solution.fluxes.CS)))
    
biomassr_pFBA = []
lacdflux_pFBA = []
maxtca_pFBA = []
mintca_pFBA = []
co2flux_pFBA = []

for i in range(-200, 200, 2):
    model = cobra.io.load_json_model("iYS854_reduced.json")
    set_boundaries(model)
    model.reactions.EX_lac__L_e.bounds = (i, i+ 1 )
    solution = cobra.flux_analysis.pfba(model)
#     solution = model.optimize()
    biomassr_pFBA.append(solution.fluxes.BIOMASS_iYS_wild_type)
    lacdflux_pFBA.append(solution.fluxes.EX_lac__L_e)
    co2flux_pFBA.append(solution.fluxes.EX_co2_e)
    maxtca_pFBA.append(max(np.abs(solution.fluxes.PDH) \
        , np.abs(solution.fluxes.OSUCCL) , np.abs(solution.fluxes.AKGDH) , np.abs(solution.fluxes.SUCOAS) \
        , np.abs(solution.fluxes.SUCD2) , np.abs(solution.fluxes.MDH3_1) , np.abs(solution.fluxes.FUM) \
        , np.abs(solution.fluxes.CS)))
    mintca_pFBA.append(min(np.abs(solution.fluxes.PDH) \
        , np.abs(solution.fluxes.OSUCCL) , np.abs(solution.fluxes.AKGDH) , np.abs(solution.fluxes.SUCOAS) \
        , np.abs(solution.fluxes.SUCD2) , np.abs(solution.fluxes.MDH3_1) , np.abs(solution.fluxes.FUM) \
        , np.abs(solution.fluxes.CS)))
```

```python
plt.plot(lacdflux_FBA, biomassr_FBA, c = 'LightBlue', label = 'FBA', linewidth=3.5)
plt.plot(lacdflux_pFBA, biomassr_pFBA, '-.', c = 'Blue', label = 'pFBA', linewidth=3.5)
plt.legend()
plt.ylim(13,15)
plt.ylabel('biomass reaction', fontsize = 16)
plt.xlabel('lactate interchange', fontsize = 16)
plt.tight_layout()
plt.show()
```
